```python
prompt = """
你是一个富有创意和童心的绘本故事撰写者，可以根据所提供的书籍内容，整理并创作出一篇兼具趣味和叙事性的小故事，用于吸引读者的注意力，让他们\\
对书籍内容产生兴趣。在你编写故事时，需要注意保证知识的准确性，不可以夸大其词、扭曲事实，字数也不能太多，\\
注意，你的目的是通过故事讲述书籍内容，而不是完全复述书籍内容，也不能忽视书里的内容。\\
你的故事必须是叙事性的，需要有主要情节、主要人物和主要事件。\\
请根据提供的书籍内容，创作一篇绘本故事：{reference}
"""
```

```python
import pandas as pd

split_contents = pd.read_csv("data/split_contents.csv")
split_contents
```

```python
# import asyncio
from erniebot_agent.chat_models import ERNIEBot
from erniebot_agent.memory import HumanMessage, AIMessage, SystemMessage, FunctionMessage

async def generate_story(reference):
    model = ERNIEBot(model = "ernie-3.5")
    # instructions = SystemMessage(story_telling_prompt)
    message = [HumanMessage(prompt.format(reference = reference))]
    #print(message)
    # ai_message = await model.chat(messages=message,SystemMessage = instructions, stream=True)
    ai_message = await model.chat(messages=message, stream=False)
    return ai_message.content
```

```python
reference = split_contents['title'][1] + split_contents['content'][1]
story = await generate_story(reference)
print(story)
```

```python
stories = []
for idx, row in split_contents.iterrows():
    print(idx)
    reference = str(row['title']) + str(row['content'])
    new_story = await generate_story(reference)
    print(new_story[:50])
    stories.append(new_story)
```
